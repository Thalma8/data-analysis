**FEATURE ENGINEERING.**


To make this data more useful, we can create new columns that extract the relevant information from the variable column and represent it in a more suitable format.


To do this, we would need to perform some string manipulation and conditional logic to extract the relevant information from the variable column and assign it to the appropriate new columns. Once we have created these new features, we can use them as input variables in a machine learning model or other statistical analysis to predict or explain some outcome of interest.

```python
def get_gender(row):
    if 'Female' in row['variable'] or '(F)' in row['variable']:
        return 'Female'
    elif 'Male' in row['variable'] or '(M)' in row['variable']:
        return 'Male'
    else:
        return 'undefined'
def get_type(row):
    if 'Currently on ART' in row['variable']:
        return 'Old'
    elif 'MOH 731   On ART' in row['variable']:
        return 'New'
    else:
        return 'Undefined'

def get_time(row):
    if 'January 2018' in  row['variable'] or '<1' in  row['variable']:
        return '01'
    elif 'February 2018' in row['variable']:
        return '02'
    elif 'March 2018' in row['variable']:
        return '03'
    elif 'April 2018' in row['variable']:
        return '04'
    elif 'May 2018' in row['variable']:
        return '05'
    elif 'June 2018' in row['variable']:
        return '06'
    elif 'July 2018' in row['variable']:
        return '07'
    elif 'August 2018' in row['variable']:
        return '08'
    elif 'September 2018' in row['variable']:
        return '09'
    elif 'October 2018' in row['variable']:
        return '10'
    elif 'November 2018' in row['variable']:
        return '11'
    elif 'December 2018' in row['variable']:
        return '12'

def get_age(row):
    if 'below 1 year' in  row['variable'] or '<1' in  row['variable']:
        return '<1'
    elif 'above 15 years' in row['variable']:
        return '>15'
    elif 'Above 15 years' in row['variable']:
        return '>15'
    elif 'Below 15 years' in row['variable']:
        return '<15'
    elif 'below 15 years' in row['variable']:
        return '<15'
    elif '1-9' in row['variable']:
        return '1-9'
    elif '10-14' in row['variable']:
        return '10-14'
    elif '15-19' in row['variable']:
        return '15-19'
    elif '20-24' in row['variable']:
        return '20-24'
    elif '25+' in row['variable']:
        return '>25'
    else:
        return 'Undefined'
def process_row(row):
    gender = get_gender(row)
    type = get_type(row)
    age= get_age(row)
    time = get_time(row)
    return pd.Series({'gender': gender, 'type': type,'age': age ,'time': time})
d49['variable'] = d49['variable'].str.strip()  # remove extra spaces in data column
d49[['gender', 'type','age', 'time']] = d49.apply(process_row, axis=1)  # create a new column 'gender'

d48['variable'] = d48['variable'].str.strip()  # remove extra spaces in data column
d48[['gender', 'type','age', 'time']] = d48.apply(process_row, axis=1)  # create a new column 'gender'

d47['variable'] = d47['variable'].str.strip()  # remove extra spaces in data column
d47[['gender', 'type','age', 'time']] = d47.apply(process_row, axis=1)  # create a new column 'gender'

d46['variable'] = d46['variable'].str.strip()  # remove extra spaces in data column
d46[['gender', 'type','age', 'time']] = d46.apply(process_row, axis=1)  # create a new column 'gender'
# Now the 'gender' column will contain the gender based on the data in each row
d49.head()
```
